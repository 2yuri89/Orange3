```python
products = df['itemDescription'].unique()
products
```

```python
one_hot=pd.get_dummies(df['itemDescription'], dtype=int)
one_hot
```

```python
df2 = df.drop('itemDescription', axis=1)
df2 = df2.join(one_hot)
df2.head()
```

```python
transaction = df2.groupby(["Member_number", "Date"])[products[:]].apply(sum)
transaction = transaction.reset_index("Member_number")
transaction
```

```python
transaction.describe()
```

```python
transaction2=transaction.iloc[:,1:]     # 고객 번호를 제외한 항몸별 거래 데이터만 저장
transaction2[transaction2 >= 1]=1   # 트랜잭션의 속성값이 1 이상인 경우 1로 대체
transaction2
```

```python
transaction2.describe()
```

```python
transaction2.to_csv('groceries_transaction2.csv')
```

```python
from mlxtend.frequent_patterns import fpgrowth, association_rules
```

```python
frequent_itemsets=fpgrowth(transaction2, min_support=0.01, max_len=3, use_colnames=True)
print('첫 10개: \n', frequent_itemsets.sort_values(by=['support'], ascending=True).head(10))
print('\n마지막 10개: \n', frequent_itemsets.sort_values(by=['support'], ascending=True).tail(10))
```

```python
frequent_itemsets.shape
```

```python
frequent_itemsets=fpgrowth(transaction2, min_support=0.001, max_len=3, use_colnames=True)
print('상위 10개:\n',frequent_itemsets.sort_values(by=['support'], ascending=True).head(10))
print('\n하위 10개:\n',frequent_itemsets.sort_values(by=['support'], ascending=True).tail(10))
```

```python
frequent_itemsets.shape
```

```python
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=0.001)
rules.sort_values(by=['lift'], ascending=False)
```

```python
fp_rules=rules.loc[:, ['antecedents', 'consequents', 'support', 'confidence', 'lift']]
fp_rules.sort_values(by=['lift'], ascending=False)
```

```python
import matplotlib.pyplot as ply
import seaborn as sns
sns.scatterplot(data=fp_rules, x='support', y='confidence', color='blue', alpha=0.6)
```

```python
sns.scatterplot(data=fp_rules, x='support', y='lift', color='green', alpha=0.6)
plt.axhline(y=1, color='red')
plt.show()
```
